Cargamos el dataframe

```python
df = pd.read_csv('dataset.csv')
```

Limpieza y transformación de datos:

- Identificar y manejar valores faltantes en diferentes tipos de datos (numéricos, textuales, fechas) utilizando estrategias adecuadas.
- Transformar variables categóricas a formato numérico (por ejemplo, mediante one-hot encoding) y normalizar o estandarizar variables numéricas según sea necesario.

```python
#miro los valores nulos
df.isnull().sum()
```

```python
# edito los valores nulos de age (Variable númerica) con la media.
df['Age'] = df['Age'].fillna(df['Age'].mean())
df.isnull().sum()
```

```python
# para variables categorigas, para embarked reemplazo con la moda
#  y elimino la fila de cabin porque tiene demasiados datos faltantes.

df['Embarked']= df['Embarked'].fillna(df['Embarked'].mode()[0])
df = df.dropna(subset=['Cabin'])
df.isnull().sum()
```

 Transformación de datos

- Convierte las variables categóricas a formato numérico utilizando técnicas como el one-hot encoding para facilitar el análisis convirtiendo variables categóricas en columnas binarias (dummies), lo cual es necesario para futuros modelados.
- Normaliza o estandariza las variables numéricas si es necesario, asegurando que todas las variables estén en una escala adecuada para futuros modelos de machine learning.

```python
# Name, Sex, Ticket, Cabin, Embarked son variables categoricas, las convierto a numéricas.
#primero veo los valores unicos
print(df['Sex'].unique())
print(df['Ticket'].unique())
print(df['Cabin'].unique())
print(df['Embarked'].unique())
```

```python
# Hago one-Hot-Encoding para Sex y Embarked, y elimino las columnas de Name, Ticket y Cabin porque no aportan información relevante para el modelo.
df = pd.get_dummies(df, columns=['Sex','Embarked'], dtype=int)
df = df.drop(columns=['Name', 'Ticket', 'Cabin'])
df.info()
```

Me guarda el nuevo dataframe 

```python
df.to_csv('dataset_cleaned.csv', index=False)
```
